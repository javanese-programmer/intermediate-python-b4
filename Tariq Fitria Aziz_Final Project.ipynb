### Step 5: Get Rid of Missing Data

```python
df.isna().sum()
```

```python
df.shape
```

1. Drop rows of data with missing values at column 'name'

```python
df.dropna(axis = 0, subset =['name', 'host_name'], inplace = True)
df.isna().sum()
```

```python
df.head()
```

```python
df.shape
```

2. Replace missing data at column 'review_per_month' by its mean

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean())
df.isna().sum()
```

```python
df.shape
```

3. Replace missing data at column 'last_review' with most common date

```python
df['last_review'] = df['last_review'].fillna(df['last_review'].mode()[0])
df.isna().sum()
```

```python
df.shape
```

### Step 6: Make a new category based on price

```python
df['price'].describe()
```

```python
df[['expensiveness']] = np.where(df["price"]>175, "Expensive", np.where(df['price']>106, 'Medium', 'Cheap'))
df.head()
```

### Step 7: Filter All Data with 'last_review' above 2019

### Step 8: Visualize the price to compare price of available room

```python
df['name'].value_counts()
```

```python
df['room_type'].value_counts()
```

```python
df_home_apt= df.loc[df['room_type'] == 'Entire home/apt']
df_home_apt.head()
```

```python
df_home_apt['price'].mean()
```

```python
df_private_room= df.loc[df['room_type'] == 'Private room']
df_private_room.head()
```

```python
df_shared_room= df.loc[df['room_type'] == 'Shared room']
df_shared_room.head()
```

```python
room_type = ['Entire Home/Apt', 'Private Room', 'Shared Room']
average_price = [df_home_apt['price'].mean(), df_private_room['price'].mean(), df_shared_room['price'].mean()]
```

```python
plt.bar(room_type, average_price, color='b')

plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.title("Room Type vs Average Price Comparison")

plt.tight_layout()
plt.show
```
